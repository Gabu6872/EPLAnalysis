```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# generates the charts and plots below the data
%matplotlib inline
```

```python
# function to check total number of null values
def check_null_values(EPL_df):
    null_values = EPL_df.isnull().sum()
    print(null_values)
    
check_null_values(EPL_df)
```

```python
# creates new columns
EPL_df['MinsPerMatch'] = (EPL_df['Mins']/EPL_df['Matches']).astype(int)
EPL_df['GoalsPerMatch'] = (EPL_df['Goals']/EPL_df['Matches']).astype(float)
EPL_df
```

```python
# calculate total goals
totalGoals = EPL_df['Goals'].sum()
totalGoals
```

```python
# penalty goals
Penalty_Goals = EPL_df['Penalty_Goals'].sum()
print(f"Out of the {totalGoals} total goals scored, {Penalty_Goals} were as a result of penalty goals")
```

```python
# penalty attempts
Penalty_Attempts = EPL_df['Penalty_Attempted'].sum()
Penalty_Attempts
```

```python
# pie chart for penalty missed vs. scored
plt.figure(figsize=(13, 6))
Penalty_not_scored = Penalty_Attempts-Penalty_Goals
data = [Penalty_Goals, Penalty_not_scored]
labels = ['Goals', 'Missed']
color = sns.color_palette("bright", 2)
plt.pie(data, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90, colors=color)
plt.axis('equal')
plt.title('Penalty Missed vs. Scored')
plt.show()
```

```python
# Players based on the group age
# Subsetting a column for players age

under_20 = EPL_df[EPL_df['Age'] <= 20]
age_between_20_and_25 = EPL_df[(EPL_df['Age'] >=20) & (EPL_df['Age'] <= 25)]
age_between_25_and_30 = EPL_df[(EPL_df['Age'] >=25) & (EPL_df['Age'] <= 30)]
above_30 = EPL_df[EPL_df['Age'] >=30]
```

```python
# plotting a pie chart
x = np.array([under_20['Name'].count(), age_between_20_and_25['Name'].count(),
             age_between_25_and_30['Name'].count(), above_30['Name'].count()])
myLabels = ["Under 20", "Between 20 and 30", "Between 25 and 30", "Above 30"]
myExplode = [0,0,0.1,0]
plt.title("Total players with age groups", fontsize=20)
plt.pie(x, labels=myLabels,colors=sns.color_palette("magma"), autopct="%.1f%%", explode=myExplode, shadow=True)
plt.show()
```

```python
# Total assists from each club
Assists_by_club = pd.DataFrame(EPL_df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Assists', data=Assists_by_club.sort_values(by='Assists', ascending=False), palette='Set2')
ax.set_xlabel('Club', fontsize=25)
ax.set_ylabel('Assists', fontsize=25)
plt.xticks(rotation=85)
plt.title('Assists by Club', fontsize=25)
plt.show()
```

```python
# Top 10 assists players
top_10_assists = EPL_df[[ 'Name', 'Club', 'Assists', 'Matches']].nlargest(n=10, columns='Assists')
top_10_assists
```

```python
# Total goals per club
goals_per_club = pd.DataFrame(EPL_df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Goals', data=goals_per_club.sort_values(by='Goals', ascending=False), palette='rocket')
ax.set_xlabel('Club', fontsize=25)
ax.set_ylabel('Goals', fontsize=25)
plt.xticks(rotation=85)
plt.title('Goals per club', fontsize=25)
plt.show()
```
